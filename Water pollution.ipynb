**Pollution Index**

```python
standards = {
    'B.O.D. (mg/l) Mean': 3,
    'Nitrate-N + Nitrite-N (mg/l) Mean': 1,
    'pH Min': 6.5,
    'pH Max': 8.5,
    'Conductivity': 1500,
    'Faecal Coli form (MPN/100ml)': 0,
    'Total Coli form (MPN/100ml)' : 0
}
```

**Pollution index for BOD**

```python
df['PI_BOD'] = df['B.O.D. (mg/l) Mean'] / standards['B.O.D. (mg/l) Mean']
```

**Pollution index for pH**

```python
def ph_index(x):
    if x < standards['pH Min']:
        return standards['pH Min'] / x
    elif x > standards['pH Max']:
        return x / standards['pH Max']
    else:
        return 1

df['PI_pH'] = df['pH Mean'].apply(ph_index)
```

**Pollution index for nitrate**

```python
df['PI_Nitrate'] = df['Nitrate-N + Nitrite-N (mg/l) Mean'] / standards['Nitrate-N + Nitrite-N (mg/l) Mean']
```

**Pollution index for conductivity**

```python
df['PI_conductivity'] = df[' Conductivity (µmhos/cm) Mean'] / standards['Conductivity']
```

**Overall pollution index (ph,BOD,Nitrate,conductivity)**

```python
df['Overall_PI'] = (df['PI_BOD']+df['PI_pH']+df['PI_Nitrate']+df['PI_conductivity'])/4
```

**Pollution level category**

```python
def pollution_level(x):
    if x < 1:
        return "Low_Pollution"
    elif x < 2:
        return "Moderate_Pollution"
    else:
        return "High_Pollution"

df['Pollution_Level'] = df['Overall_PI'].apply(pollution_level)
```

**Water quality status(BOD)**

```python
def categorize_bod(x):
    if x <= 3:
        return 'Low Polluted'
    elif x <= 10:
        return 'Moderately polluted'
    else:
        return 'Highly polluted'

df['BOD_Status'] = df['B.O.D. (mg/l) Mean'].apply(categorize_bod)
```

**Water quality status(pH)**

```python
def ph_status(x):
    if 6.5 <= x <= 8.5:
        return "Safe"
    else:
        return "Unsafe"

df['pH_Status'] = df['pH Mean'].apply(ph_status)
```

**water quality status(conductivity)**

```python
def conductivity_status(x):
    if x <= 500:
        return "Low polluted"
    elif x <= 1000:
      return "slightly polluted"
    elif x <= 1500:
      return "Moderately Polluted"
    else:
      return "Highly Polluted"
df['Water_Quality_Status'] = df[' Conductivity (µmhos/cm) Mean'].apply(conductivity_status)
```

**Water quality status(nitrate)**

```python
def Nitrate_status(x):
  if x <= 1:
    return "Low Polluted"
  elif x <= 10:
    return "slightly polluted"
  elif x <= 45:
    return "moderately polluted"
  else:
    return "Highly polluted"
df['Water_Quality_Status'] = df['Nitrate-N + Nitrite-N (mg/l) Mean'].apply(Nitrate_status)
```

**water quality status(Pollution index)**

```python
def water_quality_status(x):
    if x < 1:
        return "Low Polluted "
    elif x < 2:
        return "slightly polluted "
    elif x < 3:
        return "moderately polluted "
    else:
        return "Highly polluted"

df['Water_Quality_Status'] = df['Overall_PI'].apply(water_quality_status)
```

**Top 15 States by Average Temperature**

```python
import matplotlib.pyplot as plt
import seaborn as sns

state_temp = df.groupby('State Name')['Temperature ⁰C Mean'].mean().sort_values(ascending=False).head(15).reset_index()
plt.figure(figsize=(12, 6))
sns.barplot(data=state_temp, x='Temperature ⁰C Mean', y='State Name', hue='State Name', legend=False, palette='viridis')
plt.title('Top 15 States by Average Temperature')
plt.xlabel('Average Temperature (°C)')
plt.ylabel('State')
plt.tight_layout()
plt.savefig('avg_temp_by_state.png')
```

**Weather Stability Categories**

```python
def categorize_weather(temp_mean):
    if temp_mean < 15:
        return 'Cold'
    elif 15 <= temp_mean < 25:
        return 'Moderate'
    else:
        return 'Warm'

df['weather_category'] = df['Temperature ⁰C Mean'].apply(categorize_weather)

category_counts = df['weather_category'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', colors=['skyblue', 'orange', 'red'])
plt.title('Distribution of Weather Stability Categories')
plt.savefig('weather_category_dist.png')
```

**BOD Trend (Top 5 Polluted States)**
